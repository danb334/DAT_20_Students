```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bank_data=pd.read_csv("/Users/cvos/DAT_20_NYC/Data/bank.csv",sep=';')
```

```python
bank_data.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
job=bank_data.job.unique()
marital=bank_data.marital.unique()
education=bank_data.education.unique()
default=bank_data.default.unique()
housing=bank_data.default.unique()
loan=bank_data.loan.unique()
contact=bank_data.contact.unique()
month=bank_data.month.unique()
campaign=bank_data.campaign.unique()
pdays= bank_data.pdays.unique()
poutcome=bank_data.poutcome.unique()
y=bank_data.y.unique()

features_to_encode=['job','marital','education','default','housing','loan','contact','month','campaign','poutcome','y']

le=LabelEncoder()
job_encoder = le.fit(job)
job_encoded = le.transform(bank_data['job'])
marital_encoder = le.fit(marital)
marital_encoded = le.transform(bank_data['marital'])
education_encoder = le.fit(education)
education_encoded = le.transform(bank_data['education'])
default_encoder = le.fit(default)
default_encoded = le.transform(bank_data['default'])
housing_encoder = le.fit(housing)
housing_encoded = le.transform(bank_data['housing'])
loan_encoder = le.fit(loan)
loan_encoded = le.transform(bank_data['loan'])
contact_encoder = le.fit(contact)
contact_encoded = le.transform(bank_data['contact'])
month_encoder = le.fit(month)
month_encoded = le.transform(bank_data['month'])
campaign_encoder = le.fit(campaign)
campaign_encoded = le.transform(bank_data['campaign'])
poutcome_encoder = le.fit(poutcome)
poutcome_encoded = le.transform(bank_data['poutcome'])
y_encoder = le.fit(y)
y_encoded = le.transform(bank_data['y'])
encoded_features=[job_encoded,
                  marital_encoded,
                  education_encoded,
                  default_encoded,
                  housing_encoded,
                  loan_encoded,
                  contact_encoded,
                  month_encoded,
                  campaign_encoded,
                  poutcome_encoded,
                  y_encoded]
```

```python
bank_data_remove_cat_features=bank_data.drop(features_to_encode, axis=1)
bank_data_remove_cat_features
bank_data_features_encoded=bank_data_remove_cat_features

for i in range(len(features_to_encode)):
    bank_data_features_encoded[features_to_encode[i]]=encoded_features[i]
bank_data_features_encoded.head(10)
```

```python
features = bank_data_features_encoded.columns.tolist()
features.remove('y')

X = bank_data_features_encoded[features]
y = bank_data_features_encoded.y

#from sklearn.cross_validation import train_test_split
#X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=.3)


from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(criterion='entropy')
treeclf.fit(X,y)
```

```python
feature_importance_df = pd.DataFrame({'feature':features, 'importance':treeclf.feature_importances_})

feature_importance_df.sort(columns='importance',ascending=False)
```

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(treeclf,X,y,cv=5)
print scores
print scores.mean()
```
