```python
def findsauces (saucer, surveyor):
    sumsqerr = 0
    for i in range(0,len(saucer)):
        sumsqerr += np.power(saucer[i] - surveyor[i],2)
    return sumsqerr

def finalsauces (surveyor2):
    hotsauce2 = pd.DataFrame(hotsauce)
    #change indexes for different classifications
    #excluding dietary restrictions
    if surveyor2[18] == 1:
        hotsauce3 = hotsauce2[hotsauce2['dietary_vegan']==1]
        if hotsauce3['Product ID'].count() < hotsauce2['Product ID'].count():
            hotsauce2 = hotsauce3.reset_index()
    if surveyor2[19] == 1:
        hotsauce3 = hotsauce2[hotsauce2['dietary_noGluten']==1]
        if hotsauce3['Product ID'].count() < hotsauce2['Product ID'].count():
            hotsauce2 = hotsauce3.reset_index()
    if surveyor2[20] == 1:
        hotsauce3 = hotsauce2[hotsauce2['dietary_noAlcohol']==1]
        if hotsauce3['Product ID'].count() < hotsauce2['Product ID'].count():
            hotsauce2 = hotsauce3.reset_index()
    scores = [[] for i in range(hotsauce2['Product ID'].count())]
    for n in np.arange(1,hotsauce2['Product ID'].count()):
        arr1 = (hotsauce2.ix[n][['heat_mild','heat_medium','heat_hot','heat_hotter','heat_superhot','flavor_sweet','flavor_smoke','flavor_fruit','flavor_vinegar','flavor_salt','flavor_garlic','pepper_Jalapeno','pepper_Serrano','pepper_Scotch Bonnet','pepper_Habanero','pepper_Ghost Pepper','pepper_Moruga Scorpion','pepper_Carolina Reaper']]).values
        arr2 = surveyor2
        scores[n].append(findsauces(arr1,arr2))
        scores[n].append(hotsauce2.ix[n]['Product ID'])
    scores = filter(bool,scores)
    return sorted(scores, reverse=False)
```
