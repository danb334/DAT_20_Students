```python
fn = pd.read_csv('bank.csv', sep=';')

bank = pd.DataFrame(fn)

bank.isnull().sum()
```

### 1. There are 10 variables with string values and 7 variables with numeric values. Certain string variables are binary, such as marital, default, housing, and loan - exluding the 'y' target variable. Continuous variables include age, balance, duration, and compaign. I'm unsure of what pdays and previous represent...

```python
# Identify binary variables and handle separately from Label Encoder

for col in bank.columns:
    print str(col) + ': ' + str(len(bank[col].unique()))
```

```python
bank_data = bank.copy()

y_n_cols = ['default', 'housing', 'loan', 'y']

for col in y_n_cols:
    bank_data[col] = bank[col].map({'no': 0, 'yes': 1})
    
bank_data['marital'] = bank['marital'].map({'divorced': 0, 'single': 1, 'married':  2})

bank_data.head(10)
```

```python
# Now we will take care of the other categorical variables with LabelEncoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label_cols = ['job', 'education', 'contact', 'poutcome']

for col in label_cols:
    en_col = pd.DataFrame(le.fit_transform(bank[col]))
    bank_data[col] = en_col.values
    
# encode ordinal values for month feature

print bank_data['month'].unique()

months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
month_labels = dict(zip(months, range(1,13)))

bank_data['month'] = bank['month'].map(month_labels)

bank_data.head(10)
```

### Decision-Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split

treeclf = DecisionTreeClassifier(criterion='entropy', random_state=10)

X = bank_data.ix[:, :-1]
y = bank_data.ix[:, -1:]

print X.shape
print y.shape

X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)

treeclf.fit(X_train, y_train)

feature_importance = pd.DataFrame({'feature': X.columns, 'importance': treeclf.feature_importances_}).sort('importance', ascending=False)

feature_importance.reset_index(drop=True)

feature_importance.index = feature_importance['feature']

feature_importance = feature_importance.drop('feature', axis=1)

feature_importance
```

```python
plt.figure()

plt.bar(range(0, 16), feature_importance['importance'], color='b')
plt.xticks(range(0, 16), feature_importance.index, rotation=45)
```
