```python
#Use Alcohol, Malic Acid, Ash, Hue Proline
#Try useing f_regression (its categorical so less valuable here)
#Try covariance matrix pd.cov()

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import sklearn

from matplotlib.colors import ListedColormap
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn import neighbors, datasets, feature_selection

%matplotlib inline

url = '/Users/EVC/GA/upstream/DAT_20_Students/EVCinNYC/data/wine.data'
columns = ('class','alcohol','malicacid','ash','alcalinity','magnesium','phenols','flavanoids','nonflavanoids', 'proanthocyanins','color_intense','hue','diluted','proline')
wine = pd.read_csv(url, header=None, names=columns)

wine.head()

#ALWAYS be sure to check for NAs
# wine.isnull().any    isnull - gives logical matrix; 
#isnull.any - if any value at all, then would return True
```

```python
winedata = wine[['alcohol', 'malicacid','ash','hue','proline']]
winedata.head()
```

```python
#scaling - is important in distance models like KNN
#otherwise might bias against small scale ranges vs large ranges
winedata = sklearn.preprocessing.scale(winedata)

#had to reskin as a Data Frame again after scaling
winedata = pd.DataFrame(winedata, columns=('alcohol','malicacid','ash','hue','proline'))
winedata.head()
```

```python
wineclass = wine['class']
wineclass.head()
```

```python
print wineclass.shape, winedata.shape
```

##### Reviewing Correlation

```python
#looking at the correlation between the different features
#correlation between two items = covariance / std dev x1 * std dev x2
#correlation coefficient is inclusive of 1 to -1
# aka Pearson's R or Pearson's Correlation Coefficient
# R^2 = Pearson's R ^ 2
# 1 has line that is upwards to the right
# -1 has line that is downwards from left to right

winedata.corr()
```

```python
plt.imshow(winedata.corr(), interpolation='nearest')
```

```python
pd.scatter_matrix(winedata, figsize=(15,15), s=150, c=wineclass, alpha=0.6, cmap='brg', edgecolors='gray');
```

```python

X_train, X_test, y_train, y_test = train_test_split(winedata, wineclass, test_size=0.7, random_state = 8)
```

##### Iterating over different range of neighbors

```python


scores = []
n_neighbors = range(1,15,1)
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train,y_train)
    scores.append(clf.score(X_test,y_test))
    
plt.plot(n_neighbors, scores, linewidth=3.0)
plt.xlabel="n_neighbors"
plt.ylabel="score"
```

##### Cross Validataion

```python
from sklearn.datasets import load_iris
from sklearn.cross_validation import cross_val_score

clf = neighbors.KNeighborsClassifier(5, weights="uniform")
clf.fit(winedata, wineclass)

scores = cross_val_score(clf, winedata, wineclass, cv=5)
print scores
print scores.mean()
```

```python
clf2 = neighbors.KNeighborsClassifier(5, weights="distance")
clf2.fit(winedata, wineclass)

scores2 = cross_val_score(clf2, winedata, wineclass, cv=5)
print scores2
print scores2.mean()
```

##### Mesh graph

```python
clf = neighbors.KNeighborsClassifier(11, weights='uniform')

clf.fit(winedata, wineclass)


h = 0.02

cmap_light = ListedColormap(['#FFAAAA', "#AAFFAA", "#AAAAFF"])
cmap_bold = ListedColormap(['#FF0000', "#00FF00", "#0000FF"])

x_min, y_min = winedata.min()[['alcohol', 'proline']]
x_max, y_max = winedata.max()[['alcohol', 'proline']]
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

print xx
print yy

print xx.shape
print yy.shape


Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

Z = Z.reshape(xx.shape)
plt.figure(figsize=(18,6))

plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
```

##### Confusion Matrix

```python
#Predict 1 Predict 2 Predict 3 is columns
#Actual 1, Actual 2, Actual 3 is rows
#So can see here where had a couple wrong predictions

from sklearn import metrics

y_pred = clf.predict(X_test)

print metrics.confusion_matrix(y_pred, y_test)

conf_matrix = metrics.confusion_matrix(y_pred, y_test)
```

```python
pd.DataFrame(conf_matrix, columns=['Pred1','Pred2','Pred3'], index=['Actual1', 'Actual2','Actual3'])
```

##### Compute the Precision, Recall and F1 Score

```python
TruePos = conf_matrix[0,0] + conf_matrix[1,1] + conf_matrix[2,2]
FalsePos = conf_matrix[0,1] + conf_matrix[0,2] + conf_matrix[1,0] + conf_matrix[1,2] + conf_matrix[2,0] + conf_matrix[2,1]
FalseNeg = conf_matrix[0,1] + conf_matrix[0,2] + conf_matrix[1,0] + conf_matrix[1,2] + conf_matrix[2,0] + conf_matrix[2,1]

Precision = TruePos / float(TruePos+FalsePos)
Recall = TruePos / float(TruePos + FalseNeg)

F1_Score = 2 * (Precision * Recall)/(Precision + Recall)

print TruePos
print "FalsePos:", FalsePos
print "FalseNeg:", FalseNeg
print "Precision:", Precision
print "Recall:", Recall
print "F1 Score:", F1_Score
```

##### Which features most interesting?

```python
#could use F-regression - but more for regression probs and this is a classification problem
#F-regression would pretty much show what did above in correlation
#Could drop out one feature at a time, but time-consuming
#recursive feature elimination algorithm
#tries best feature, then best two features, then best three features, etc.

from sklearn.svm import SVC
from sklearn.feature_selection import RFECV
from sklearn.cross_validation import StratifiedKFold
 

svc = SVC(kernel='linear')

rfecv = RFECV(estimator=svc, step=1, cv=StratifiedKFold(y_test, 2), scoring='accuracy')

rfecv.fit(winedata,wineclass)

print rfecv.n_features_

#TRY TO COMPARE TO ANTHONY'S PYTHON NOTEBOOK
```
