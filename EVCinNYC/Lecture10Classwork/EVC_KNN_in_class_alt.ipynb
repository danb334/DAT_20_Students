```python
%matplotlib inline
from __future__ import division
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import load_iris  #pre-packaged set - categorize feature of plant based on params

from matplotlib.colors import ListedColormap
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn import neighbors, datasets, feature_selection
#neighbors module is the KNN
iris = datasets.load_iris()
  #datasets is a sklearn function - not really a pandas function
    #dir(iris) shows all the attributes -- __class__ are built in functions 
    #have to load into a data frame to work with it

iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=12)
```

```python
scores = []

n_neighbors = range(1,80,1)   #
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n, weights='distance')    #(n) = # of K
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
    
plt.plot(n_neighbors, scores, linewidth=3.0)
plt.xlabel="n_neighbors"
plt.ylabel="score"
```

```python
#trying to speed type what was done in class
scores2 = []
for k in range (1,5):
    clf = neighbors.KNeighborsClassifier(k, weights='uniform')
    clf.fit(X_train,y_train)
    score = clf.score(X_test, y_test)
    
    scores2.append(score)
    
plt.plot(range(1,5),scores)
```
